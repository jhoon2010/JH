```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
tol    = 10**(-5)
A1, A2 = 0, 1
Bsol   = 0
err    = 100

def int(A):
    x    = -5
    xmax = 0
    y    = 0
    vy   = 0.1
    dx   = 0.001
    vy   = vy + 0.5 * (-1) * (A - np.abs(x)) * y * dx
    while(x <= xmax):
        y  = y + vy * dx
        vy = vy + (-1) * (A - np.abs(x)) * y * dx
        x = x + dx
    return y

def intplot(A):
    x    = -5
    xmax = 5
    y    = 0
    vy   = 0.1
    dx   = 0.001
    vy   = vy + 0.5 * (-1) * (A - np.abs(x)) * y * dx
    xlist, ylist = [], []
    while(x <= xmax):
        xlist.append(x)
        ylist.append(y)
        y  = y + vy * dx
        vy = vy + 0.5 * (-1) * (A - np.abs(x)) * y * dx
        x = x + dx
    return xlist, ylist

def find_eigenvalue(A1,A2):
    tol    = 10**(-5)
    Bsol   = 0
    err    = 100

    while(np.abs(err) > tol):
        B1  = int(A1)
        print(B1)
        B2  = int(A2)
        err = Bsol - B2
        if(np.abs(err) > tol):
            yp = A2 + (A2 - A1)/(B2 - B1) * err
            A1 = A2
            A2 = yp
        else:
            print(yp)

    xlist, ylist = intplot(yp)
    plt.plot(xlist, ylist)
```

```python
find_eigenvalue(0,1)
```
