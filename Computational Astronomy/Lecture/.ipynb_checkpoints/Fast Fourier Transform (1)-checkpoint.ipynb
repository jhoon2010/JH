# Fast Fourier Transform (1)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 8
Omg = np.zeros((n,n), dtype = complex)
for j in range(n):
    for i in range(n):
        W = np.exp(-2 * np.pi * complex(0,1) / n)
        Omg[j][i] = pow(W, i*j)
```

```python
from scipy import linalg
x = np.arange(n) / float(n)
f = np.cos(2 * np.pi * x)

ft = Omg.dot(f)
print(ft.real)
print(ft.imag)

t = np.linspace(0,1,100)
plt.plot(t,np.cos(2 * np.pi * t))
t = np.linspace(0,1,9)
plt.scatter(t,np.cos(2 * np.pi * t))
```

```python
fsamp = 8
t = np.arange(0,1,1/fsamp)
f = np.sin(2 * np.pi * t)
F = np.fft.fft(f)
plt.plot(F.real,'ro--',label='real')
plt.plot(F.imag,'bo--',label='imag')
```

```python
A1, A2 = 2, 1
freq1, freq2 = 10, 50
fsamp = 256
t = np.arange(0,1,1/fsamp)
n = len(t)
f = A1 * np.sin(2 * np.pi * freq1 * t) + A2 * np.cos(2 * np.pi * freq2 * t)
f += 0.2 * np.random.randn(n)

F = np.fft.fft(f)
freq = np.fft.fftfreq(n,1/fsamp)
F_shifted = np.fft.fftshift(F)
freq_shifted = np.fft.fftshift(freq)

plt.figure(figsize=(5,15))
plt.subplot(3,1,1)
plt.plot(t, f)
plt.subplot(3,1,2)
plt.plot(F.real,'r')
plt.plot(F.imag,'b')
plt.subplot(3,1,3)
plt.plot(F_shifted.real,'r')
plt.plot(F_shifted.imag,'b')
```
