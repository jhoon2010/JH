# Monte Carlo Method 3

```python
(1-2.35)/(100**(1-2.35)-1)
```

```python
a=2.35
C=(1-a)/(100**(1-a)-1**(1-a))
C
```

```python
x=np.random.rand(10000)
a=2.35
result=[]
for i in range(0,10000):
    y=(1+(1-a)/C*(x[i]))**(1/(1-a))
    result.append(y)
    
y
```

```python
result
```

```python
bins=np.arange(1,100,1)
plt.hist(result, bins, log=True,normed=1,histtype='step')
plt.gca().set_xscale("log")
```
