```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

# ODE 3

```python
def deriv(y,t):
    dydt0 = 2*y[0] - y[0]*y[1]
    dydt1 = -9*y[1] + 3.*y[0]*y[1]
    return [dydt0,dydt1]

yini = np.array([1.5,2.0])
t=np.linspace(0,10,1000)
y=odeint(deriv,yini,t)
u=y[:,0]
v=y[:,1]

plt.plot(t,u)
plt.plot(t,v)
```

```python
def func(x,y):
    dydt0=(2-y[1])*y[0]
    dydt1=(-9+3*y[0])*y[1]
    return np.array([dydt0,dydt1])

def rk4_ODEset(x,y,h):
    hh=0.5*h
    v1=h*func(x,y)
    v2=h*func(x+hh,y+0.5*v1)
    v3=h*func(x+hh,y+0.5*v2)
    v4=h*func(x+h,y+v3)
    return y+(v1+2*v2+2*v3+v4)/6

t,u,v=[0],[1.5],[2.0]

x=0
y=[1.5,2.0]
h=0.001

for i in range (10000):
    y=rk4_ODEset(x,y,h)
    x+=h
    u=np.append(u,y[0])
    v=np.append(v,y[1])
    t=np.append(t,x)
    
plt.plot(t,u)
plt.plot(t,v)
```

```python
plt.plot(u,v)
plt.xlim(0,6)
```

```python
def deriv(y,t):
    g=10
    l=1
    dydt0 = y[1]
    dydt1 = -(g/l)*np.sin(y[0])
    return [dydt0,dydt1]

yini = np.array([np.pi/100,0])
t=np.linspace(0,10,1000)
y=odeint(deriv,yini,t)
u=y[:,0]
v=y[:,1]

plt.plot(t,u)
plt.plot(t,v)
```

```python
def deriv(y,t,gam,omega):
    dydt0=y[1]
    dydt1=-gam*y[1]-omega*omega*y[0]
    return [dydt0,dydt1]

yini=[0,2]
gam=0.4
omega=2
y=odeint(deriv,yini,t,args=(gam,omega))
u=y[:,0]
v=y[:,1]
plt.plot(u,v)
```
