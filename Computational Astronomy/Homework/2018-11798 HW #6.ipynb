## Problem 1

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
from scipy import signal
```

```python
def f(x):
    return 100*(x[1] - x[0]**2)**2+(1-x[0])**2
```

```python
def gradf(x):
    return np.array([-400*x[0]*(x[1]-x[0]**2)-2*(1-x[0]),200*(x[1]-x[0]**2)])
```

```python
def golden2D(x, dr, TOL):
    R,err,loop=0.61803399,10,-1 
    adr = np.sqrt(dr**[0] + dr[1]**2)
    a= x-dr/adr
    b= x+dr/adr 
    while(err > TOL): 
        loop += 1 
        x1= b-R*(b-a) 
        x2= a+R*(b-a) 
        f1=f(x1) 
        f2=f(x2) 
        if(f2>f1): b=x2 
        else: a=x1 
        err = sum((a-b)**2)
        err = np.sqrt(err)
    xmin=a 
    return xmin, err, loop
```

```python
plt.rcParams["figure.figsize"] = (12,9)
x = np.linspace(-2.5,1.5,300)
y = np.linspace(-0.5,2.5,300)
X, Y = np.meshgrid(x,y)
E = f([X, Y])
cmap = plt.cm.gist_heat
dmax = np.max(E)
dmin = np.min(E)
con = plt.contourf(x, y, E, levels = np.linspace(dmin, dmax, 50),cmap = "hsv",alpha = 0.7)

TOL = 10**(-12)
x = np.array([-2,2])
dr1 = gradf(x)
plt.plot(x[0], x[1], 'ro')
x1, err, loop = golden2D(x, dr1, TOL)
err = (x1[1] - x[1])**2 + (x1[0] - x[0])**2
plt.plot([x[0], x1[0]],[x[1], x1[1]], 'b')
plt.plot(x1[0], x1[1], 'ro');x=x1

cnt = 0
while (err > TOL):
    cnt = cnt + 1
    dr2 = (-1) * gradf(x) + ((gradf(x)[0]**2 + gradf(x)[1]**2) / (dr1[0]**2 + dr1[1]**2)) * dr1
    x1, err, loop = golden2D(x,dr2,TOL)
    dr1 = dr2
    err = (x1[1] - x[1])**2 + (x1[0] - x[0])**2
    plt.plot((x[0], x1[0]),[x[1], x1[1]], 'b')
    plt.plot(x1[0], x1[1], 'ro')
    x = x1
plt.show()
print('%d iterations'%cnt)
print((x[0], x[1]))
```

## Problem 2

```python
def f(x):
    return 100 * (x[1] - x[0]**2)**2 + (1 - x[0])**2 + 100 * (x[2] - x[1]**2)**2 + (1 - x[2])**2
```

```python
def gradf(x):
    return np.array([(-400) * x[0] * (x[1] - x[0]**2) - 2 * (1 - x[0]), 200 * (x[1] - x[0]**2) - 400 * x[1] * (x[2] - x[1]**2), 200 * (x[2] - x[1]**2) - 2 * (1 - x[2])])
```

```python
def golden3D(x,dr,TOL):
    R, err, loop = 0.61803399,10.,-1
    adr = np.sqrt(dr[0]**2 + dr[1]**2 + dr[2]**2)
    a = x - (dr / adr)
    b = x + (dr / adr)
    while (err > TOL):
        loop = loop + 1
        x1 = b - R * (b - a)
        x2 = a + R*(b - a)
        f1 = f(x1)
        f2 = f(x2)
        if(f2>f1): 
            b = x2
        else: 
            a = x1
        err = sum((a - b)**2)
        err = np.sqrt(err)
    xmin=a
    return xmin,err,loop
```

```python
TOL = 10**(-12)
x = np.array([0,2,1])
dr1 = gradf(x)
x1, err, loop = golden3D(x, dr1, TOL)
err=(x1[2] - x[2])**2 + (x1[1] - x[1])**2 + (x1[0] - x[0])**2


cnt= 0
while (err>TOL):
    cnt = cnt + 1
    dr2 = gradf(x) + ((gradf(x)[0]**2 + gradf(x)[1]**2 + gradf(x)[2]**2) / (dr1[0]**2 + dr1[1]**2 + dr1[2]**2)) * dr1
    x1, err, loop = golden3D(x, dr2, TOL)
    dr1 = dr2
    err = (x1[2] - x[2])**2 + (x1[1] - x[1])**2 + (x1[0] - x[0])**2
    x = x1
print('minimum point : ({},{},{})'.format(x[0],x[1],x[2]),'\n','%d iterations'%cnt)
```

## Problem 3

```python
# (a)
fn = 'sol_vel.dat'
t, vH, vCa = np.loadtxt(fn,unpack = True, usecols = [0,1,2])
t = t * 60
plt.plot(t, vH, 'r')
plt.plot(t, vCa, 'b')
```

```python
# (b)
F1 = np.fft.fft(vH)
F1 = np.fft.fftshift(F1)

F2 = np.fft.fft(vCa)
F2 = np.fft.fftshift(F2)

n = len(t)
k = np.fft.fftfreq(n, 1 / n)
k = np.fft.fftshift(k)

plt.figure(1, figsize = (12,6))
plt.plot(k, F1.real,'b')
plt.plot(k, F2.real,'r')
plt.xlabel('freuquency(1/s)')

plt.figure(2, figsize = (12,6))
plt.plot(k, F1.imag,'b')
plt.plot(k, F2.imag,'r')
plt.xlabel('frequency(1/s)')
```

```python
# (c) H - alpha peak frequency
abs_real1, abs_imag1 = (np.abs(F1.real)).tolist(),(np.abs(F1.imag)).tolist()
print('cos component : %.6f'%abs(k[abs_real1.index(max(abs_real1))])
     ,'sin component : %.6f'%abs(k[abs_imag1.index(max(abs_imag1))]))
```

```python
# (c) Ca II peak frequency
abs_real2, abs_imag2 = (np.abs(F2.real)).tolist(),(np.abs(F2.imag)).tolist()
print('cos component : %.6f'%abs(k[abs_real2.index(max(abs_real2))])
     ,'sin component : %.6f'%abs(k[abs_imag2.index(max(abs_imag2))]))
```

```python
# 따라서 H-alpha 의 주기
print('{}s'.format(np.amax(t) / abs(k[abs_real1.index(max(abs_real1))])))
```

```python
# 따라서 Ca II 의 주기
print('{}s'.format(np.amax(t) / abs(k[abs_real2.index(max(abs_real2))])))
```
