```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(taumax):
    i , reflected , transmitted = 0 , 0 , 0
    plot_x = []
    plot_z = []
    # step 1
    zeta1  = np.random.rand(1000)
    zeta2  = np.random.rand(1000)
    zeta3  = np.random.rand(1000)
    while(i < 1000):
        x , y , z , zmax = 0 , 0 , 0 , 1
        plot_x = [0]
        plot_z = [0]

        mu     = np.sqrt(zeta1[i])
        theta  = np.arccos(mu)
        phi    = 2*np.pi*zeta2[i]

        dir_x  = np.sin(theta) * np.cos(phi)
        dir_y  = np.sin(theta) * np.sin(phi)
        dir_z  = np.cos(theta)

        while(1):
            # step 2 and step 3
            tau    = -np.log(zeta3[i])
            L      = tau*zmax/taumax
            x      = x + L*dir_x
            y      = y + L*dir_y
            z      = z + L*dir_z
            if(i == 999):
                plot_x.append(x)
                plot_z.append(z)
            # step 4
            if(z < 0):
                reflected+=1
                break
            if(z > zmax):
                transmitted+=1
                break
            # step 5
            zeta4  = np.random.rand(1)
            zeta5  = np.random.rand(1)

            mu     = 2*zeta5 - 1
            theta  = np.arccos(mu)
            phi    = 2*np.pi*zeta4

            dir_x  = np.sin(theta) * np.cos(phi)
            dir_y  = np.sin(theta) * np.sin(phi)
            dir_z  = np.cos(theta)
        i+=1
    
    if(taumax < 1):
        plt.plot(plot_x,plot_z)
        plt.scatter(plot_x,plot_z,s=100)
        plt.ylim(0,1)
    if(taumax == 1):
        plt.plot(plot_x,plot_z)
        plt.scatter(plot_x,plot_z,s=10)
        plt.ylim(0,1)
    if(taumax > 1):
        plt.plot(plot_x,plot_z)
        plt.scatter(plot_x,plot_z,s=10)
        plt.ylim(0,0.3)
    return reflected/1000, transmitted/1000 # 반사 확률 / 투과 확률 출력
```

```python
f(0.01)
```

```python
f(0.1)
```

```python
f(1) # 투과한 경우
```

```python
f(1) # 반사된 경우
```

```python
f(10)
```

### taumax 가 0.01, 0.1 처럼 작은 값일 때는 상호작용 확률이 낮고, 평균 자유행로가 길어서 투과 확률이 반사 확률에 비해 훨씬 크다. 그러나 taumax가 1일 때는, 투과 확률과 반사 확률이 반반 정도로 나타난다. 그리고 taumax가 1보다 훨씬 큰 10의 값을 가질 때에는 많은 상호작용을 해서 대부분이 반사되는 것을 알 수 있다.
