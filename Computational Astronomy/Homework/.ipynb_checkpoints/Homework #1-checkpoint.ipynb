# Problem 1

```python
def f():
    x=int(input("Input first integer")) # 입력
    y=int(input("Input second integer")) #
    if(x>=y): # 첫번째 정수가 더 큰 경우
        q=int(x/y) # 몫은 x/y 의 정수부분
        r=x%y # 나머지 연산
    else: # 두번째 정수가 더 큰 경우
        q=int(y/x)
        r=y%x 
    print('two integers: {}    and {}'.format(x,y)) #출력
    print('quotient    : {0:d}'.format(q))
    print('remainder   : {}'.format(r))
```

# Problem 3

```python
nmax=int(input("Input maximum integer"))

def p(n):
    for i in range(2,int(np.sqrt(n)+1)): # range 이므로 sqrt n 에 1을 더해줘야
        if n%i==0: 
            return 0 # 합성수라면 0 반환
    return 1 # 소수라면 1 반환

count=0
k=2 # 2부터 nmax 까지 이기 때문
while(k<=nmax):
    if (p(k)==1): # 소수라면 count 에 1 더함
        count+=1
    k+=1
print(count) # 여태까지 세어진 소수의 수
            
        
```

# Problem 4

```python
fn='hw1_p4.dat'
x,a,b=np.loadtxt(fn,unpack=True,usecols=[0,1,2])
```

```python
file=open('hw1_p4_fourth.dat','w')
i=3 # 위에서부터 4번째 줄
while (i<=9999):
    txt=str(x[i])+'\t'+str(a[i])+'\t'+str(b[i])+'\n'
    file.write(txt)
    i+=4 # 4번째 줄마다

file.close()
```

```python
plt.plot(x,a*x,'-',label='t=t1')
plt.plot(x,b*x,':',label='t=t2')
plt.plot(x,a*b*x,'--',label='t=t3')
plt.legend()
plt.xlabel('time',fontsize=20)
plt.ylabel('value',fontsize=20)
plt.show()
```
