```python
# 3차원에서
result = []
for i in range(100): # 100번 반복
    cnt = 0 # 범위 안 점의 개수를 세기 위해서
    x = np.random.rand(1000) # 좌표
    y = np.random.rand(1000) # 좌표
    z = np.random.rand(1000) # 좌표
    for j in range(1000):
        if((x[j]**2 + y[j]**2 + z[j]**2) < 1): # 3차원 구 내부이면
            cnt+=1 # cnt +1 
    result.append(cnt)

print(np.average(result)*8/1000) # 100번 반복한 결과
print(np.std(result)/125) # 표준편차
print(np.pi**1.5/(0.75*(np.pi)**0.5)) # 참값
```

# Problem 2

```python
def f(taumax):
    i , reflected , transmitted = 0 , 0 , 0
    plot_x = []
    plot_z = []
    # step 1
    zeta1  = np.random.rand(1000)
    zeta2  = np.random.rand(1000)
    zeta3  = np.random.rand(1000)
    while(i < 1000):
        x , y , z , zmax = 0 , 0 , 0 , 1
        plot_x = [0]
        plot_z = [0]

        mu     = np.sqrt(zeta1[i])
        theta  = np.arccos(mu)
        phi    = 2*np.pi*zeta2[i]

        dir_x  = np.sin(theta) * np.cos(phi)
        dir_y  = np.sin(theta) * np.sin(phi)
        dir_z  = np.cos(theta)

        while(1):
            # step 2 and step 3
            tau    = -np.log(zeta3[i])
            L      = tau*zmax/taumax
            x      = x + L*dir_x
            y      = y + L*dir_y
            z      = z + L*dir_z
            if(i == 999):
                plot_x.append(x)
                plot_z.append(z)
            # step 4
            if(z < 0):
                reflected+=1
                break
            if(z > zmax):
                transmitted+=1
                break
            # step 5
            zeta4  = np.random.rand(1)
            zeta5  = np.random.rand(1)

            mu     = 2*zeta5 - 1
            theta  = np.arccos(mu)
            phi    = 2*np.pi*zeta4

            dir_x  = np.sin(theta) * np.cos(phi)
            dir_y  = np.sin(theta) * np.sin(phi)
            dir_z  = np.cos(theta)
        i+=1
    
    if(taumax < 1):
        plt.plot(plot_x,plot_z)
        plt.scatter(plot_x,plot_z,s=100)
        plt.ylim(0,1)
    if(taumax == 1):
        plt.plot(plot_x,plot_z)
        plt.scatter(plot_x,plot_z,s=10)
        plt.ylim(0,1)
    if(taumax > 1):
        plt.plot(plot_x,plot_z)
        plt.scatter(plot_x,plot_z,s=10)
        plt.ylim(0,0.3)
    return reflected/1000, transmitted/1000 # 반사 확률 / 투과 확률 출력
```

### taumax 가 0.01, 0.1 처럼 작은 값일 때는 상호작용 확률이 낮고, 평균 자유행로가 길어서 투과 확률이 반사 확률에 비해 훨씬 크다. 그러나 taumax가 1일 때는, 투과 확률과 반사 확률이 반반 정도로 나타난다. 그리고 taumax가 1보다 훨씬 큰 10의 값을 가질 때에는 많은 상호작용을 해서 대부분이 반사되는 것을 알 수 있다.

```python
def d(x, u): # 도함수
    return (2/x-1/x**2)/(u-1/u)


def rk4(f, x0, xn, y0, h): # rk 4 method
    xlist, ylist = [], [] # 값을 저장할 list
    x, y = x0, y0 # 초기값
    while np.abs(y-1)>0.01: # 정의 안되는 부분을 제외하고
        xlist.append(x)
        ylist.append(y)
        k1=f(x,y)
        k2=f(x-0.5*h, y-0.5*h*k1)
        k3=f(x-0.5*h, y-0.5*h*k2)
        k4=f(x-h,y-h*k3)
        y-=h/6*(k1+2*k2+2*k3+k4)
        x-=h
    while x<=xn: # x 값을 다시 키우면서 계산
        xlist.append(x)
        ylist.append(y)
        k1=f(x,y)
        k2=f(x+0.5*h, y+0.5*h*k1)
        k3=f(x+0.5*h, y+0.5*h*k2)
        k4=f(x+h,y+h*k3)
        y+=h/6*(k1+2*k2+2*k3+k4)
        x+=h
    return np.array(xlist), np.array(ylist)
```

```python
def g(u): # 일반적인 경우
    x = 5
    h = 0.001
    u = u + 0.5*h*(2/x-1/x**2)/(u-1/u)
    xlist, ylist = [], []
    while(x >= 0.1):
        xlist.append(x)
        ylist.append(u)
        u = u - h*(2/x-1/x**2)/(u-1/u)
        x = x - h
    return np.array(xlist), np.array(ylist)
```
