```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def d(x, u): # 도함수
    return (2/x-1/x**2)/(u-1/u)


def rk4(f, x0, xn, y0, h): # rk 4 method
    xlist, ylist = [], [] # 값을 저장할 list
    x, y = x0, y0 # 초기값
    while np.abs(y-1)>0.01: # 정의 안되는 부분을 제외하고
        xlist.append(x)
        ylist.append(y)
        k1=f(x,y)
        k2=f(x-0.5*h, y-0.5*h*k1)
        k3=f(x-0.5*h, y-0.5*h*k2)
        k4=f(x-h,y-h*k3)
        y-=h/6*(k1+2*k2+2*k3+k4)
        x-=h
    while x<=xn: # x 값을 다시 키우면서 계산
        xlist.append(x)
        ylist.append(y)
        k1=f(x,y)
        k2=f(x+0.5*h, y+0.5*h*k1)
        k3=f(x+0.5*h, y+0.5*h*k2)
        k4=f(x+h,y+h*k3)
        y+=h/6*(k1+2*k2+2*k3+k4)
        x+=h
    return np.array(xlist), np.array(ylist)
```

```python
def g(u): # 일반적인 경우
    x = 5
    h = 0.001
    u = u + 0.5*h*(2/x-1/x**2)/(u-1/u)
    xlist, ylist = [], []
    while(x >= 0.1):
        xlist.append(x)
        ylist.append(u)
        u = u - h*(2/x-1/x**2)/(u-1/u)
        x = x - h
    return np.array(xlist), np.array(ylist)
```

```python
x1, u1=g(3) # x=5 에서 u=3 일때
```

```python
plt.plot(x1, u1)
```
