## Problem 1

```python
def V(E):
    return np.exp(-np.sqrt(2 / E) - E * 0.2) - 0.1

# 왜냐하면, V = 0.1 일 때 E의 값을 찾아야 하므로
```

```python
# 1 - (a)
t = np.linspace(-10,20,500)
plt.plot(t, V(t))
# 대략적인 그래프의 개형을 파악하였음. 이를 토대로 root finding 진행
```

```python
# peak 좌측의 근 찾기
TOL = 10**(-10)
err = 1
a, b = 0.01 , 2.5
cnt  = 1

while(err > TOL):
    Va = V(a)
    Vb = V(b)
    c  = (a + b) * 0.5
    Vc = V(c)
    err = np.abs(b - c)
    if(err < TOL):
        print(c)
    if(Va * Vc > 0):
        a = c
    else:
        b = c
    cnt = cnt + 1
```

```python
# peak 우측의 근 찾기
TOL = 10**(-10)
err = 1
a, b = 2.5, 10
cnt  = 1

while(err > TOL):
    Va = V(a)
    Vb = V(b)
    c  = (a + b) * 0.5
    Vc = V(c)
    err = np.abs(b - c)
    if(err < TOL):
        print(c)
    if(Va * Vc > 0):
        a = c
    else:
        b = c
    cnt = cnt + 1
```

```python
# 따라서 두 근은 각각 0.40525062, 9.17899793 임을 알 수 있다.
```

```python
# 1 - (b)
def dV(E):
    return np.exp(-0.2 * E - np.sqrt(2 / E)) * ((1 / E)**(1.5) / np.sqrt(2) - 0.2)

# V(E) 를 미분한 값이 0 이 되는 지점을 찾으면 된다.
```

```python
# 그래프의 개형을 대략 알기 때문에, 구간을 정할 수 있다.
TOL = 10**(-10)
err = 1
a, b = 0.5, 4
cnt  = 1

while(err > TOL):
    dVa = dV(a)
    dVb = dV(b)
    c  = (a + b) * 0.5
    dVc = dV(c)
    err = np.abs(b - c)
    if(err < TOL):
        print(c)
    if(dVa * dVc > 0):
        a = c
    else:
        b = c
    cnt = cnt + 1
```

```python
# 따라서 V를 최대로 만드는 E_max 값은 2.3207944 임을 알 수 있다.
```

## Problem 3

```python
fn = 'elnino.dat'
dP = np.loadtxt(fn, unpack=True)
```

```python
# 3 - (a)
t = np.linspace(0,168,168)
plt.plot(t,dP)
plt.xlabel('time(months)')
plt.ylabel('delta P')
```

```python
# 3 - (b)
dP_fft = np.fft.fft(dP)
dP_fft = np.fft.fftshift(dP_fft)
                         
t_fft  = np.fft.fftfreq(len(t))
t_fft  = np.fft.fftshift(t_fft)

plt.semilogy(t_fft,np.absolute(dP_fft)**2)
plt.xlim(0,0.5)
plt.xlabel('Frequency')
plt.ylabel('Power')
```

```python
# 3 - (c)
t_fft_pos = t_fft[t_fft > 0]
P_dP = 1 / (t_fft_pos[np.argmax(np.absolute(dP_fft[t_fft > 0])**2)])
print(P_dP)
```

```python
# 따라서 가장 dominant 한 주기는 12개월이다.
```

```python
# 지구 상의 서로 다른 두 지점에 대한 기압차이인데, 1년을 단위로
# 지구가 태양 주위를 공전하고, 기상 현상도 이에 따라 주기적으로
# 나타난다. 따라서 1년이 가장 지배적인 주기인 점이 타당하다고 볼 수 있다.
```

```python
# 3 - (d)
t_fft_second = t_fft[(t_fft < 0.07) & (t_fft > 0)]
P_dP = 1 / (t_fft_second[np.argmax(np.absolute(dP_fft[(t_fft < 0.08) & (t_fft > 0)])**2)])
print(P_dP)

# power spectrum 을 보았을 때, 두 번째로 dominant 한 주기는
# 0 과 첫번째 peak 사이에 있다. 따라서 t_fft 의 범위를
# 0 과 첫번째 peak 사이로 설정한 후, 그 곳에서 최대값을 찾아서
# 전체적으로는 두 번째로 큰 peak 의 값을 골라내는 방식을 이용하였다.
```

```python
def rho(r):
    if(r == 0):
        return 0
    else:
        return np.array[1 / (4 * np.pi * r**2)]
```

```python
# 4 - (a)
r1 = np.random.rand(10000)
theta = np.arccos(2 * r1 - 1)
r2 = np.random.rand(10000)
phi = 2 * np.pi * r2
r3 = np.random.rand(10000)

x = r3 * np.sin(theta) * np.cos(phi) 
y = r3 * np.sin(theta) * np.sin(phi)
z = r3 * np.cos(theta)

plt.scatter(x, y)
```

```python
rho = np.zeros(1000)
r   = 0
dr  = 0.001
for i in range(1000):
    r = np.sqrt(x[i]**2 + y[i]**2 + z[i]**2)
    order = math.floor(r / dr)
    rho[order] = rho[order] + 0.0001
```

```python
t = np.linspace(0,1,1000)
plt.plot(t,rho)
plt.xlabel('radius')
plt.ylabel('density')
```

```python
# 주어진 식에 의한 density profile 은 r 제곱에 반비례하기 때문에
# 바깥쪽의 밀도가 매우 낮은 것을 확인할 수 있다.
# 반면 random 하게 입자를 뿌린 경우에는 바깥쪽의 구각일 수록,
# 많은 입자를 포함하고 있다. 그래서 바깥쪽으로 갈 수록 밀도가
# 증가하는 경향을 확인할 수 있다.
```

```python
# 4 - (b)
mi = 0.0001

r1 = np.random.rand(10000)
theta = np.arccos(2 * r1 - 1)
r2 = np.random.rand(10000)
phi = 2 * np.pi * r2

vx = np.sin(theta) * np.cos(phi) 
vy = np.sin(theta) * np.sin(phi)
vz = np.cos(theta)

K = 0
for i in range(0, 10000):
    K = K + vx[i]**2 + vy[i]**2 + vz[i]**2

print(0.5 * K * mi)
# 왜냐하면 모든 입자의 질량이 동일하기 때문에, 시그마 기호에서 빼서
# 마지막에 한 번에 계산하였다.
```

```python
# 따라서 K 값은 0.5 이다.
```

```python
W = 0
mi = 0.0001
for i in range(0,10000):
    for j in range(0, 10000):
        if(i != j):
            W = W + 1 / np.sqrt((x[i] - x[j])**2 + (y[i] - y[j])**2 + (z[i] - z[j])**2)

W = (-0.5) * W * mi * mi
# 역시 마지막에 한 번에 계산함
print(W)
```
