```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.subplot(321)
t=np.arange(0,50,0.01)
r=1/9
w=r*np.sqrt(0.96)
d=np.arctan(2*w*0.2/(1-w**2))
plt.plot(t,(-1)*np.exp(-0.2*t)*np.cos(np.sqrt(1-0.2**2)*t),'r-.') # x_c
plt.plot(t,1/np.sqrt((1-w**2)**2+4*w**2*0.04)*np.cos(w*t-d),'b--') #x_p
plt.plot(t,(-1)*np.exp(-0.2*t)*np.cos(np.sqrt(1-0.2**2)*t)+1/np.sqrt((1-w**2)**2+4*w**2*0.04)*np.cos(w*t-d),'g')
plt.xlim(0,30)
plt.ylim(-2,2)
fig=plt.figure(1,figsize=(4,4))
plt.title('ratio=1/9')
plt.subplots_adjust(left=0.08, right=0.95, top=0.99, bottom=0.1, wspace=0.25,hspace=0.45)


plt.subplot(322)
t=np.arange(0,50,0.01)
r=1/3
w=r*np.sqrt(0.96)
d=np.arctan(2*w*0.2/(1-w**2))
plt.plot(t,(-1)*np.exp(-0.2*t)*np.cos(np.sqrt(1-0.2**2)*t),'r-.') # x_c
plt.plot(t,1/np.sqrt((1-w**2)**2+4*w**2*0.04)*np.cos(w*t-d),'b--') #x_p
plt.plot(t,(-1)*np.exp(-0.2*t)*np.cos(np.sqrt(1-0.2**2)*t)+1/np.sqrt((1-w**2)**2+4*w**2*0.04)*np.cos(w*t-d),'g')
plt.xlim(0,30)
plt.ylim(-2,2)
fig=plt.figure(1,figsize=(4,4))
plt.title('ratio=1/3')
plt.subplots_adjust(left=0.08, right=0.95, top=0.99, bottom=0.1, wspace=0.25,hspace=0.45)



plt.subplot(323)
t=np.arange(0,50,0.01)
r=1.1
w=r*np.sqrt(0.96)
d=np.arctan(2*w*0.2/(1-w**2))
plt.plot(t,(-1)*np.exp(-0.2*t)*np.cos(np.sqrt(1-0.2**2)*t),'r-.') # x_c
plt.plot(t,1/np.sqrt((1-w**2)**2+4*w**2*0.04)*np.cos(w*t-d),'b--') #x_p
plt.plot(t,(-1)*np.exp(-0.2*t)*np.cos(np.sqrt(1-0.2**2)*t)+1/np.sqrt((1-w**2)**2+4*w**2*0.04)*np.cos(w*t-d),'g')
plt.xlim(0,30)
plt.ylim(-2.5,2.5)
fig=plt.figure(1,figsize=(4,4))
plt.title('ratio=1.1')
plt.subplots_adjust(left=0.08, right=0.95, top=0.99, bottom=0.1, wspace=0.25,hspace=0.45)


plt.subplot(324)
t=np.arange(0,50,0.01)
r=3
w=r*np.sqrt(0.96)
d=np.arctan(2*w*0.2/(1-w**2))
plt.plot(t,(-1)*np.exp(-0.2*t)*np.cos(np.sqrt(1-0.2**2)*t),'r-.') # x_c
plt.plot(t,1/np.sqrt((1-w**2)**2+4*w**2*0.04)*np.cos(w*t-d),'b--') #x_p
plt.plot(t,(-1)*np.exp(-0.2*t)*np.cos(np.sqrt(1-0.2**2)*t)+1/np.sqrt((1-w**2)**2+4*w**2*0.04)*np.cos(w*t-d),'g')
plt.xlim(0,30)
plt.ylim(-1.2,0.7)
fig=plt.figure(1,figsize=(4,4))
plt.title('ratio=3')
plt.subplots_adjust(left=0.08, right=0.95, top=0.99, bottom=0.1, wspace=0.25,hspace=0.45)

plt.subplot(325)
t=np.arange(0,50,0.01)
r=6
w=r*np.sqrt(0.96)
d=np.arctan(2*w*0.2/(1-w**2))
plt.plot(t,(-1)*np.exp(-0.2*t)*np.cos(np.sqrt(1-0.2**2)*t),'r-.') # x_c
plt.plot(t,1/np.sqrt((1-w**2)**2+4*w**2*0.04)*np.cos(w*t-d),'b--') #x_p
plt.plot(t,(-1)*np.exp(-0.2*t)*np.cos(np.sqrt(1-0.2**2)*t)+1/np.sqrt((1-w**2)**2+4*w**2*0.04)*np.cos(w*t-d),'g')
plt.xlim(0,30)
plt.ylim(-2,0.7)
fig=plt.figure(1,figsize=(4,4))
plt.title('ratio=6, A=1m/s^2')
plt.subplots_adjust(left=0.08, right=0.95, top=0.99, bottom=0.1, wspace=0.25,hspace=0.45)


plt.subplot(326)
t=np.arange(0,50,0.01)
r=6
w=r*np.sqrt(0.96)
d=np.arctan(2*w*0.2/(1-w**2))
plt.plot(t,(-1)*np.exp(-0.2*t)*np.cos(np.sqrt(1-0.2**2)*t),'r-.') # x_c
plt.plot(t,20/np.sqrt((1-w**2)**2+4*(w**2)*0.04)*np.cos(w*t-d),'b--') #x_p
plt.plot(t,(-1)*np.exp(-0.2*t)*np.cos(np.sqrt(1-0.2**2)*t)+20/np.sqrt((1-w**2)**2+4*(w**2)*0.04)*np.cos(w*t-d),'g')
plt.xlim(0,30)
plt.ylim(-1.5,1.2)
fig=plt.figure(1,figsize=(4,4))
plt.title('ratio=6, A=20m/s^2')
plt.subplots_adjust(left=0.08, right=0.95, top=0.99, bottom=0.1, wspace=0.25,hspace=0.45)
```
