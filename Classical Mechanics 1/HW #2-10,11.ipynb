```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bi(a,b,h,x1,x2): # alpha 의 범위 / step / 세로 축 범위
    s=[]
    r=[]
    fig = plt.figure(figsize=(7.2,4.05))
    while(a<=b):
        x=0.5
        for t in range(0,500):
            y=a*x*(1-x)
            if(t>=400): # 400회 반복이면 초기 조건 없어졌다고 판단함
# 충분히 진행되었을 때(=초기 조건 영향이 없어졌을 때) 값들을 list 에 저장
                s.append(a)
                r.append(y)
            x=y
        a+=h
    plt.scatter(s,r,s=0.003)
    plt.ylim(x1,x2)
```

```python
bi(2.8,4,0.2,0,1) # 2.8 ~ 4.0 / alpha 0.2 씩 증가
```

```python
bi(2.8,4,0.001,0,1) # 2.8 ~ 4.0 / alpha 0.001 씩 증가
```

```python
bi(3.840,3.857,0.000005,0.44,0.55) # 3.840 ~ 3.857 / alpha 0.000005 씩 증가
```
