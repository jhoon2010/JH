```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(x,y,B):
    rx=[x] # 초기 값
    ry=[y] # 초기 값
    t=0
    dt=10**(-4) # time step 설정
    n=0
    for i in range(0,10000000): # euler method 로 미방 풀이
        t+=dt
        x=x+y*dt
        y=y+(-0.1*y-x**3+B*np.cos(t))*dt
        if(i==(int((2*n*np.pi)//dt))):
            # 일정한 간격(2 pi 초마다) 의 점들만 plot 하기 위함
            rx.append(x)
            ry.append(y)
            n+=1
        i+=1
    plt.scatter(rx,ry,s=0.1)
```

```python
plot(0,0,9.8) # 아래 모두 초기 위치 (0,0) 으로 동일 , k = 9.8
```

```python
plot(0,0,10) # k = 10
```

```python
plot(0,0,12) # k = 12
```
