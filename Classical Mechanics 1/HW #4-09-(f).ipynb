```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(xini):
    g = 1
    R = 1
    w = np.sqrt(2)
    tlist, xlist, xrlist = [], [], []
    t = 0
    dt = 0.001
    x, v, xreal = xini * np.pi / 180, 0, xini * np.pi / 180
    wc = np.sqrt(g / R)
    w = np.arccos(wc**2 / w**2)
    v = v - 0.5 * (w**4 - wc**2) / w**2 * x * dt

    while(t <= 10):
        xlist.append(x)
        xrlist.append(xreal)
        tlist.append(t)
        x = x + v * dt
        xreal = (np.pi) * xini / 180 * np.cos((np.sqrt(w**4 - wc**2))/w * t)
        v = v - (w**4 - wc**2) / w**2 * x * dt
        t = t + dt
    
    plt.subplot(2,1,1)
    plt.plot(tlist, xlist,'r')
    plt.plot(tlist, xrlist,'b')
    
    err = np.array(xlist)-np.array(xrlist)
    plt.subplot(2,1,2)
    plt.semilogy(tlist,np.abs(err))
```

```python
f(1)
```

```python
f(10)
```
