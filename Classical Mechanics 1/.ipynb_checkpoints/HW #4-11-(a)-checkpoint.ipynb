```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = np.linspace(0.5,10,1000)
V = - 2 / 3 * r**(-1.5) + 0.5 / r**2
plt.plot(r,V)
```

```python
def V(r):
    return - 2 / 3 * r**(-1.5) + 0.5 / r**2

def f(r):
    h = 0.001
    return (V(r + h) - V(r)) / h
```

```python
err = 1
TOL = 10**(-3)
r = 0.5
dr = 0.01
while(err > TOL):
    if(np.abs(f(r)) < TOL):
        print(r)
        break
    else:
        r = r + dr
```

```python
err = 1
TOL = 10**(-3)
r = 0.5
dr = 0.0001
while(err > TOL):
    if(np.abs(V(r) + 0.1) < TOL):
        print(r)
        break
    else:
        r = r + dr
```

```python
t = 0
dt = 0.01
r = 0.66
v = 0
u = 1 / r
v = v + 0.5 * (np.sqrt(u) - u) * dt
tlist, xlist,ylist = [], [], []
while(t <= 30):
    x = r * np.cos(t)
    y = r * np.sin(t)
    xlist.append(x)
    ylist.append(y)
    
    u = u + v * dt
    v = v + (np.sqrt(u) - u) * dt
    t = t + dt
    r = 1 / u

plt.plot(xlist, ylist)
```
