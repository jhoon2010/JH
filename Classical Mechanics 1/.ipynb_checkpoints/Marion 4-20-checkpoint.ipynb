```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot():
    rx=[0] # 초기 조건
    ry=[0] # 초기 조건
    for i in range(0,10000): # 횟수
        x=ry[i]+1-1.4*(rx[i])**2 # 주어진 관계식
        y=0.3*rx[i] # 주어진 관계식
        rx.append(x) # 값을 list 에 추가
        ry.append(y) # 위와 동일
        i+=1
    plt.figure(figsize=(7.2,5.5))
    plt.plot(rx,ry,'bo',ms=0.03)
    plt.xlim(-1.5,1.5)
    plt.ylim(-0.45,0.45)
```

```python
plot()
```
