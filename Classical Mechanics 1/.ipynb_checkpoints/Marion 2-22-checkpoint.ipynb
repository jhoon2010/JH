```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w=2
A=10
Ey=10
B=2
t=np.arange(0,10,0.001)
plt.plot(-A*np.cos(w*t)/w+Ey*t/B,A*np.sin(w*t)/w)
plt.grid()
plt.title('A>|Ey/B|')
```
