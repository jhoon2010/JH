```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y  = 0.2
vy = 0
t  = 0
dt = 0.01
a  = 0.2
b  = 1
g  = 1
w  = 0.1
rt = []
ry = []
vy = vy + 0.5 * dt * ((w * w * a / b)*np.cos(y - w * t) - g * np.sin(y) / b)
while (t <= 100):
    rt.append(t)
    ry.append(y)
    y = y + vy * dt
    vy = vy + dt * ((w * w * a / b)*np.cos(y - w * t) - g * np.sin(y) / b)
    t = t + dt

plt.plot(rt,ry)
```
