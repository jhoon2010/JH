```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rk4(f, r0, h, tend):
    pt, pr= [], []
    t, r=0, np.array(r0,dtype=float)
    
    while t<= tend:
        pt.append(t)
        pr.append(r.copy())
        
        k1=f(t,r)
        k2=f(t+h/2,r+h/2*k1)
        k3=f(t+h/2,r+h/2*k2)
        k4=f(t+h,r+h*k3)
        r+=(h/6)*(k1+2*k2+2*k3+k4)
        t+=h
    return np.array(pt), np.array(pr)

def f_damped(t,r):
    px, vx = r
    dx=vx
    ddx=-2*vx-px
    
    return np.array([dx, ddx])
```

```python
r0=[0,10]
tout,rout1=rk4(f_damped,r0,0.01,10)
```

```python
r0=[0,-10]
tout2,rout2=rk4(f_damped,r0,0.01,10)
```

```python
r0=[15,-10]
tout3,rout3=rk4(f_damped,r0,0.01,10)
```

```python
r0=[-20,-20]
tout3,rout4=rk4(f_damped,r0,0.01,10)
```

```python
plt.plot(rout1[:,0],rout1[:,1],'r')
plt.plot(rout2[:,0],rout2[:,1],'g')
plt.plot(rout3[:,0],rout3[:,1],'b')
plt.plot(rout4[:,0],rout4[:,1],'y')

x=np.arange(-40,40,0.01)
plt.plot(x,-x)
plt.xlim(-30,30)
plt.ylim(-25,25)
plt.title('phase diagram')
```
