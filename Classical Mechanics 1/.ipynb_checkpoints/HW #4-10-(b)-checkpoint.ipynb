```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(e,b):
    t = 0
    dt = 0.01
    xlist, ylist = [], []
    while(t <= 1000):
        r = 1 / (1 + e * np.cos(b * t))
        x = r * np.cos(t)
        y = r * np.sin(t)
        xlist.append(x)
        ylist.append(y)
        t = t + dt
    plt.plot(xlist,ylist)
```

```python
f(0.5,1.1)
```

```python
f(0.5,np.sqrt(2))
```

```python
f(0.5,0.9)
```

```python
f(0.5,1/np.sqrt(2))
```
