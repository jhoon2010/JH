```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.subplot(311)
t=np.arange(0,30,0.01)
d=np.pi/2
x=3*np.cos(t)
y=3*np.cos(t+d)
plt.plot(x,y,'b')

plt.subplot(312)
t=np.arange(0,30,0.01)
d=np.pi*5/6
x=3*np.cos(t)
y=3*np.cos(t+d)
plt.plot(x,y,'b')

plt.subplot(313)
t=np.arange(0,30,0.01)
d=np.pi
x=3*np.cos(t)
y=3*np.cos(t+d)
plt.plot(x,y,'b')


```

```python
plt.subplot(221)
t=np.arange(0,1000,0.01)
d=np.pi*0
r=3/4
x=3*np.sin(t)
y=3*np.sin(r*t+d)
plt.plot(x,y,'b')
plt.xlim(-4,4)
plt.ylim(-4,4)

plt.subplot(222)
t=np.arange(0,1000,0.01)
d=np.pi*0
r=2
x=3*np.sin(t)
y=3*np.sin(r*t+d)
plt.plot(x,y,'b')
plt.xlim(-4,4)
plt.ylim(-4,4)

plt.subplot(223)
t=np.arange(0,1000,0.01)
d=np.pi/3
r=2
x=3*np.cos(t)
y=3*np.cos(r*t+d)
plt.plot(x,y,'b')
plt.xlim(-4,4)
plt.ylim(-4,4)

plt.subplot(224)
t=np.arange(0,1000,0.01)
d=np.pi*0.5
r=2
x=3*np.cos(t)
y=3*np.sin(r*t+d)
plt.plot(x,y,'b')
plt.xlim(-4,4)
plt.ylim(-4,4)

```
