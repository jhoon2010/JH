```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def iteration(x,a): # 초기 값과 a 값 받음
    r=[x]
    for i in range(0,20):
        if((x>0) and (x<0.5)): # iteration 의 범위 지정
            x=2*x*a
            r.append(x) # x값을 list 에 저장
        else:
            x=2*a*(1-x)
            r.append(x) # x값을 list에 저장
        i+=1
    j=list(range(0,21))
    plt.plot(j,r,'r') # x값을 iteration 횟수에 따라 출력
    plt.plot(j,r,'ro')
    return r
    
    
```

```python
iteration(0.2,0.4) # alpha = 0.4 인 경우
```

```python
iteration(0.2,0.7) # alpha = 0.7 인 경우
```
