```python
hecx, hecy, hecz = np.loadtxt("HECTOR.txt", skiprows = 4, usecols = [2,3,4], unpack = 1)
```

```python
jupx, jupy, jupz = np.loadtxt("JUPITER.txt", skiprows = 4, usecols = [2,3,4], unpack = 1)
```

```python
print(hecz)
```

```python
fig = plt.figure()
ax = fig.gca(projection = '3d')

plt.plot(hecx,hecy,hecz,'b')
plt.plot(jupx,jupy,jupz,'g')
ax.scatter(hecx[50],hecy[50],hecz[50],'ko')
ax.scatter(jupx[50],jupy[50],jupz[50],'yo')
ax.scatter(0,0,0)
ax.set_xlabel('x(AU)')
ax.set_ylabel('y(AU)')
ax.set_zlabel('z(AU)')
```

```python
x = []
y = []
z = []

for i in range(0, 219):
    rel_x = (hecx[i] * jupx[i] + hecy[i] * jupy[i] + hecz[i] * jupz[i]) / np.sqrt(jupx[i]**2 + jupy[i]**2 + jupz[i]**2)
    
    del_y = hecy[i] - jupy[i]
    theta = np.arccos(del_y * jupy[i] / np.sqrt(jupx[i]**2 + jupy[i]**2 + jupz[i]**2) / np.abs(del_y))
    rel_y = np.abs(del_y) * np.sin(theta)
    
    del_z = hecz[i] - jupz[i]
    theta = np.arccos(del_z * jupz[i] / np.sqrt(jupx[i]**2 + jupy[i]**2 + jupz[i]**2) / np.abs(del_z))
    rel_z = del_z * np.sin(theta)
    
    x.append(rel_x)
    y.append(rel_y)
    z.append(rel_z)
```

```python
fig = plt.figure()
ax = fig.gca(projection = '3d')

plt.plot(x, y, z)
ax.set_xlabel('x(AU)')
ax.set_ylabel('y(AU)')
ax.set_zlabel('z(AU)')
```
