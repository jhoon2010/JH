```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(m1,m2,k,H):
    g  = 9.8
    v  = 0
    y  = H
    t  = 0
    dt = 0.001
    v  = v + 0.5 * dt * (g - k * v / m1)
    rt1 , ry1 , rv1 = [] , [] , []
    while(y >= 0):
        rt1.append(t)
        ry1.append(y)
        rv1.append(v)
        y = y - v * dt
        v = v + dt * (g - k * v / m1)
        t = t + dt
        if(y < 0):
            T_1_fall = t

    T_0_1 = np.sqrt(2 * H / g)
    delta_T_1 = T_1_fall - T_0_1
    delta_T_1_predict = 0.3333 * (k * H) * T_0_1 /(T_0_1 * m1 * g - k * H)
    
    print(T_0_1, T_1_fall, delta_T_1, delta_T_1_predict)
    
    g  = 9.8
    v  = 0
    y  = H
    t  = 0
    dt = 0.001
    v  = v + 0.5 * dt * (g - k * v / m2)
    rt2 , ry2 , rv2 = [] , [] , []
    while(y >= 0):
        rt2.append(t)
        ry2.append(y)
        rv2.append(v)
        y = y - v * dt
        v = v + dt * (g - k * v / m2)
        t = t + dt
        if(y < 0):
            T_2_fall = t
    
    T_0_2 = np.sqrt(2 * H / g)
    delta_T_2 = T_2_fall - T_0_2
    delta_T_2_predict = 0.3333 * (k * H) * T_0_2 /(T_0_2 * m2 * g - k * H)
    
    print(T_0_2, T_2_fall, delta_T_2, delta_T_2_predict)
    
    g  = 9.8
    v  = 0
    y  = H
    t  = 0
    dt = 0.001
    v  = v + 0.5 * dt * (g)
    rt3 , ry3 , rv3 = [] , [] , []
    while(y >= 0):
        rt3.append(t)
        ry3.append(y)
        rv3.append(v)
        y = y - v * dt
        v = v + dt * (g)
        t = t + dt
        if(y < 0):
            T_3_fall = t

    T_0_3 = np.sqrt(2 * H / g)
    
    print(T_0_3, T_3_fall)
    
    plt.subplot(1,2,1)
    plt.plot(rt1,ry1,'b')
    plt.plot(rt2,ry2,'r')
    plt.plot(rt3,ry3,'g')
    plt.ylim(0,2100)
    plt.subplot(1,2,2)
    plt.plot(rt1,rv1,'b')
    plt.plot(rt2,rv2,'r')
    plt.plot(rt3,rv3,'g')
    plt.ylim(0,200)
```

```python
f(10,1,0.1,1960)
```
