```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f():
    R  = 6370000
    r  = R
    v  = 0
    M  = 5.67 * 10**24
    G  = 6.67 * 10**(-11)
    t  = 0
    dt = 1
    tt, rt , vt, at = [], [], [], []
    a  = (-1) * G * M / R**3 * r
    v  = v + 0.5 * dt * a
    while(t <= 10000):
        tt.append(t)
        rt.append(r)
        vt.append(v)
        at.append(a)
        r = r + v * dt
        a = (-1) * G * M / R**3 * r
        v = v + dt * a
        t = t + dt
    return tt, rt, vt, at
```

```python
t, r, v, a = f()
```

```python
plt.plot(t,r)
```

```python
plt.plot(t,v)
```

```python
plt.plot(t,a)
```
