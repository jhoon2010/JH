```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fn = 'forfft.csv'
```

```python
x, y = np.loadtxt(fn, unpack=True, usecols = [0,1], delimiter = ',')
```

```python
F1 = np.fft.fft(y) / len(y)
F2 = np.fft.fftshift(y) / len(y)

x = x * (12195)

n = len(x)
k = np.fft.fftfreq(n,1/ n)
k = np.fft.fftshift(k)

#plt.subplot(1,,1)
plt.figure(1, figsize = (6, 3))
plt.plot(k,F1.real,'b')
plt.plot(k,F2.real,'r')
#plt.subplot(2,1,2)
plt.figure(2, figsize = (6, 3))
plt.plot(k, F1.imag,'b')
plt.plot(k,F2.imag,'r')

```

```python
fft_mag1 = abs(F1)
fft_mag2 = abs(F2)
```

```python
fs = len(x)
aa = np.linspace(-(fs/2),fs/2,fs)
ffff = np.fft.fftshift(fft_mag1)
```

```python
plt.stem(aa, ffff)
```
