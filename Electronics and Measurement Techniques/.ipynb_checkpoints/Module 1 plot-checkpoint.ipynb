## Thermistor

```python
# 실험군 (1 MOhm + 220 Ohm , Voltage : 2V)
t01 = np.loadtxt('DATA\Os2meandata1004-1428.txt')
t02 = np.loadtxt('DATA\Os2meandata1004-1435.txt')
t03 = np.loadtxt('DATA\Os2meandata1004-1508.txt')
t04 = np.loadtxt('DATA\Os2meandata1004-1512.txt')
t05 = np.loadtxt('DATA\Os2meandata1004-1614.txt')
t06 = np.loadtxt('DATA\Os2meandata1004-1601.txt')
t07 = np.loadtxt('DATA\Os2meandata1004-1608.txt')
t08 = np.loadtxt('DATA\Os2meandata1004-1611.txt')
t09 = np.loadtxt('DATA\Os2meandata1004-1617.txt')
t10 = np.loadtxt('DATA\Os2meandata1004-1438.txt')
t11 = np.loadtxt('DATA\Os2meandata1004-1442.txt')
t12 = np.loadtxt('DATA\Os2meandata1004-1445.txt')
t13 = np.loadtxt('DATA\Os2meandata1004-1622.txt')

# 대조군(1 MOhm 만 연결한 경우)
x01 = np.loadtxt('DATA\Os2meandata1001-2256.txt')
x02 = np.loadtxt('DATA\Os2meandata1001-2258.txt')
x03 = np.loadtxt('DATA\Os2meandata1001-2259.txt')

MEAN = (x01[10] + x02[10] + x03[10]) / 3 # 대조군 중 2V 줬을 때 평균

# 대조군과 실험군의 전압 차이
d01 = MEAN - np.mean(t01)
d02 = MEAN - np.mean(t02)
d03 = MEAN - np.mean(t03)
d04 = MEAN - np.mean(t04)
d05 = MEAN - np.mean(t05)
d06 = MEAN - np.mean(t06)
d07 = MEAN - np.mean(t07)
d08 = MEAN - np.mean(t08)
d09 = MEAN - np.mean(t09)
d10 = MEAN - np.mean(t10)
d11 = MEAN - np.mean(t11)
d12 = MEAN - np.mean(t12)
d13 = MEAN - np.mean(t13)

# 표준 편차
dev01 = np.loadtxt('DATA\Os2devdata1004-1428.txt')
dev02 = np.loadtxt('DATA\Os2devdata1004-1435.txt')
dev03 = np.loadtxt('DATA\Os2devdata1004-1508.txt')
dev04 = np.loadtxt('DATA\Os2devdata1004-1512.txt')
dev05 = np.loadtxt('DATA\Os2devdata1004-1614.txt')
dev06 = np.loadtxt('DATA\Os2devdata1004-1601.txt')
dev07 = np.loadtxt('DATA\Os2devdata1004-1608.txt')
dev08 = np.loadtxt('DATA\Os2devdata1004-1611.txt')
dev09 = np.loadtxt('DATA\Os2devdata1004-1617.txt')
dev10 = np.loadtxt('DATA\Os2devdata1004-1438.txt')
dev11 = np.loadtxt('DATA\Os2devdata1004-1442.txt')
dev12 = np.loadtxt('DATA\Os2devdata1004-1445.txt')
dev13 = np.loadtxt('DATA\Os2devdata1004-1622.txt')

a = [dev01, dev02, dev03, dev04, dev05, dev06, dev07, dev08, dev09, dev10, dev11, dev12, dev13]
dev = np.zeros(13)
for i in range(13):
    dev[i] = np.mean(a[i]) / np.sqrt(3)

d = [d03, d04, d05, d06, d07, d08, d09, d10, d11, d12, d13]
t = [t03, t04, t05, t06, t07, t08, t09, t10, t11, t12, t13]
T = [10, 20, 30, 40, 50, 60, 90, 100, 110, 120, 180]
R = np.zeros(11)
for i in range(11):
    R[i] = d[i] / np.mean(t[i]) * (10**6) # 저항 구하는 식

Rerr = np.zeros(11)
for i in range(11):
    Rerr[i] = dev[i] / np.mean(t[i]) * (10**6) # 저항의 오차

plt.errorbar(T, R, yerr = Rerr, fmt = 'bo', elinewidth = 0.5, capsize = 3, ecolor = 'k')
plt.title('Thermistor, Resistance vs. Temperature')
plt.xlabel('Temperature(℃)') # 가열 시간에 따라서 plot 해서, 온도로 바꾸려면 보정 필요함!
plt.ylabel('Resistance(Ω)')
```

## 저항 구하기 + IV Sweep

```python
# 대조군 8.2 kOhm
x01 = np.loadtxt('DATA\Os2meandata1002-0040.txt') # Voltage : (-2.5V , 2.5V , delta = 0.1V)
x02 = np.loadtxt('DATA\Os2meandata1002-0044.txt') # Voltage : (2.5V , -2.5V , delta = -0.1V) 

# 실험군(8.2 kOhm + 220 Ohm)
y01 = np.loadtxt('DATA\Os2meandata1002-0058.txt') # Voltage : (-2.5V , 2.5V , delta = 0.1V)
y02 = np.loadtxt('DATA\Os2meandata1002-0102.txt') # Voltage : (2.5V , -2.5V , delta = -0.1V)

# 실험군과 대조군 차이
d01 = x01 - y01
d02 = x02 - y02

# 걸어준 전압
Volt1 = np.loadtxt('DATA\oltdata11002-0040.txt')
Volt2 = np.loadtxt('DATA\oltdata11002-0044.txt')

# 저항 구하기
R1 = d01 / Volt1 * 8200
R2 = d02 / Volt2 * 8200

# 표준편차
dev01 = np.loadtxt('DATA\Os2devdata1002-0058.txt')
dev02 = np.loadtxt('DATA\Os2devdata1002-0102.txt')

# 저항의 오차
Rerr1 = np.zeros(51)
Rerr2 = np.zeros(51)
for i in range(51):
    Rerr1[i] = dev01[i] / x01[i] * (8200)
    Rerr2[i] = dev02[i] / x02[i] * (8200)

# print(R1)
# print(R2)
# V = 0 에서의 값이 크게 튀는 것을 확인 가능. plot 할 때 y 범위 제한

plt.figure(figsize = (10, 10))

plt.subplot(211)
plt.errorbar(Volt1, R1, yerr = Rerr1, fmt = 'bo', elinewidth = 0.5, capsize = 3, ecolor = 'k', label = 'go up')
plt.errorbar(Volt2, R2, yerr = Rerr2, fmt = 'ro', elinewidth = 0.5, capsize = 3, ecolor = 'k', label = 'go down')
plt.title('Resistance vs. Voltage')
plt.legend(loc = 'lower right')
plt.xlabel('Input Voltage (V)')
plt.ylabel('Resistance (Ω)')
plt.ylim(50, 350)

# 통계 내기 위해 특이값 제거(0 부근 값을 제외하면 std 가 더 줄어들 것으로 예상됨.)
R1 = R1.tolist()
R2 = R2.tolist()
del R1[25]
del R2[25]

print(np.mean(R1), np.std(R1))
print(np.mean(R2), np.std(R2))

# IV sweep

# 전류의 표준편차
devv01 = dev01 / 8200
devv02 = dev02 / 8200

plt.subplot(212)
plt.errorbar(d01, y01 / 8200, yerr = devv01, fmt = 'bo', elinewidth = 0.5, capsize = 3, ecolor = 'k', label = 'go up')
plt.errorbar(d02, y02 / 8200, yerr = devv02, fmt = 'ro', elinewidth = 0.5, capsize = 3, ecolor = 'k', label = 'go down')
plt.title('IV Curve')
plt.legend(loc = 'lower right')
plt.xlabel('Voltage across 220Ω (V)')
plt.ylabel('Current (A)')

# 그냥 plot하면 에러바가 거의 보이지 않아서 에러바를 100배 확대하여 plot함.
```

## Histogram + Gaussian

```python
def gauss(g,x,num):
    return num * np.exp(-0.5 * (x - g[0])**2 / g[1]**2)
```

```python
x1 = np.loadtxt('DATA\Os1meandata1004-1853.txt') # 10ms 250
x2 = np.loadtxt('DATA\Os1meandata1004-1847.txt') # 100ms 250
x3 = np.loadtxt('DATA\Os1meandata1004-1907.txt') # 1000ms 250
x4 = np.loadtxt('DATA\Os1meandata1005-1556.txt') # 1000ms 2500

y1 = np.loadtxt('DATA\Os2meandata1004-1853.txt') # 10ms 250
y2 = np.loadtxt('DATA\Os2meandata1004-1847.txt') # 100ms 250
y3 = np.loadtxt('DATA\Os2meandata1004-1907.txt') # 1000ms 250


plt.figure(figsize = (30, 20))
num_bins = 15

plt.subplot(2,2,1)
n, bins, patches = plt.hist(x1[1:], num_bins)
plt.xlabel('Voltage across 220Ω (V)')
plt.ylabel('#')
t = np.linspace(0.048, 0.054, 10000)
g = [np.mean(x1), np.std(x1)]
print(g)
plt.plot(t, gauss(g, t, 90))
plt.text(0.053, 80, '$\mu$ = 0.0516\n$\sigma$ = 0.000579\nN = 250')
plt.xlim(0.048, 0.054)

plt.subplot(2,2,2)
n, bins, patches = plt.hist(x2[1:], num_bins)
plt.xlabel('Voltage across 220Ω (V)')
plt.ylabel('#')
t = np.linspace(0.051, 0.054, 10000)
g = [np.mean(x2), np.std(x2)]
print(g)
plt.plot(t, gauss(g, t, 100))
plt.text(0.0535, 90, '$\mu$ = 0.0519\n$\sigma$ = 0.000269\nN = 250')
plt.xlim(0.051, 0.054)

plt.subplot(2,2,3)
n, bins, patches = plt.hist(x3[1:], num_bins)
plt.xlabel('Voltage across 220Ω (V)')
plt.ylabel('#')
t = np.linspace(0.050, 0.053, 10000)
g = [np.mean(x3), np.std(x3)]
print(g)
plt.plot(t, gauss(g, t, 90))
plt.text(0.0525, 78, '$\mu$ = 0.0517\n$\sigma$ = 0.000206\nN = 250')
plt.xlim(0.050, 0.053)

plt.subplot(2,2,4)
n, bins, patches = plt.hist(x4[1:], num_bins)
plt.xlabel('Voltage across 220Ω (V)')
plt.ylabel('#')
t = np.linspace(0.044, 0.054, 10000)
g = [np.mean(x4), np.std(x4)]
print(g)
plt.plot(t, gauss(g, t, 1330))
plt.text(0.0525, 1250, '$\mu$ = 0.0510\n$\sigma$ = 0.000139\nN = 2500')
plt.xlim(0.044, 0.054)
```
