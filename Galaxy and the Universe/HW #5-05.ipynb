```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = np.linspace(0, 10, 1000)
```

```python
plt.plot(d, np.exp(-7.67 * (d**0.25 - 1)))
plt.plot(d, 5.67 * np.exp(- d * 1.68))
plt.xlabel('R / Re')
plt.ylabel('I / Ie')
```

```python
plt.semilogx(d, np.exp(-7.67 * (d**0.25 - 1)))
plt.semilogx(d, 5.67 * np.exp(- d * 1.68))
plt.xlabel('log(R / Re)')
plt.ylabel('I / Ie')
plt.ylim(0, 200)
```

```python
plt.semilogy(d, np.exp(-7.67 * (d**0.25 - 1)))
plt.semilogy(d, 5.67 * np.exp(- d * 1.68))
plt.xlabel('R / Re')
plt.ylabel('log(I / Ie)')
```

```python
plt.loglog(d, np.exp(-7.67 * (d**0.25 - 1)))
plt.loglog(d, 5.67 * np.exp(- d * 1.68))
plt.xlabel('log(R / Re)')
plt.ylabel('log(I / Ie)')
```

```python
plt.semilogy(d**0.25, np.exp(-7.67 * (d**0.25 - 1)))
plt.semilogy(d**0.25, 5.67 * np.exp(- d * 1.68))
plt.xlabel('(R / Re)^(1/4)')
plt.ylabel('log(I / Ie)')
```
