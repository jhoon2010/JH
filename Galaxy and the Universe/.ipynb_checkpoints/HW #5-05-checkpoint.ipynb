```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = np.linspace(0, 10, 1000)
```

```python
plt.plot(d, np.exp(-7.67 * (d**0.25 - 1)))
plt.plot(d, np.exp(- d * 1.68))
```

```python
plt.semilogx(d, np.exp(-7.67 * (d**0.25 - 1)))
plt.semilogx(d, np.exp(- d * 1.68))
```

```python
plt.semilogy(d, np.exp(-7.67 * (d**0.25 - 1)))
plt.semilogy(d, np.exp(- d * 1.68))
```

```python
plt.loglog(d, np.exp(-7.67 * (d**0.25 - 1)))
plt.loglog(d, np.exp(- d * 1.68))
```
