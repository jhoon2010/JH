```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 은하 중심 거리 R을 입력하면, V_r vs. longitude 를 plot 해주는 함수
def f(R):
    Llist = []
    vlist = []
    dL    = 0.01
    
    # solar circle 보다 안쪽에 있는 천체에 대하여
    if(R <= 8):
        L_max = np.arcsin(R / 8) # 은경의 최대 값을 구해줌
        L = (-1) * L_max # 초기 은경을 범위 중 가장 작은 값으로 설정해 준다.
        while(L <= L_max): # 주어진 은경 범위에서, 즉 - L_max 부터 + L_max 까지
            Llist.append(L)
            vlist.append(200 * np.sin(L) * ((8/R)**(1.5) - 1)) # 주어진 식에 따라 V_r 을 구해준 후 저장
            L = L + dL
        Llist = np.array(Llist)
        Llist = Llist * 180 / np.pi # 은경 list 에 저장되어있는 값을 deg 단위로 변환
        L_max = L_max * 180 / np.pi # L_max 의 값도 rad 단위에서 deg 단위로 변환
        plt.plot(Llist, vlist)
        plt.xlim(L_max, (-1) * L_max)
        plt.xlabel("galactic longitude (deg)")
        plt.ylabel("radial velocity (km/s)")
        
    # solar circle 바깥 쪽에 있는 천체에 대하여
    else:
        L = (-1) * np.pi
        while(L <= np.pi):
            Llist.append(L)
            vlist.append(200 * np.sin(L) * ((8/R)**(1.5) - 1)) # 주어진 식에 따라 V_r 을 구해준 후 저장
            L = L + dL
        Llist = np.array(Llist)
        Llist = Llist * 180 / np.pi # 은경 list 에 저장되어있는 값을 deg 단위로 변환
        plt.plot(Llist, vlist)
        plt.xlim(180, -180)
        plt.xlabel("galactic longitude (deg)")
        plt.ylabel("radial velocity (km/s)")
```

```python
f(4) # R = 4kpc 인 경우
```

```python
f(6) # R = 6kpc 인 경우
```

```python
# 위의 두 경우, 즉 항성이 solar circle 안에 있는 경우를 분석해보면 L = L_max 일 때 시선 속도가 가장 큰 것을 볼 수 있다.
# L = L_max 일 때, 항성의 공전 속도 중 시선 속도 성분이 가장 많기 때문에 시선 속도가 가장 클 것이라고 예측할 수 있는데, 
# 이 예측과 일치하는 결과임을 알 수 있다. 또한 교재에서도 동일한 내용을 확인할 수 있다.
# 같은 방식으로 은경이 0인 지점에 있는 천체는 시선 속도가 0이고 접선 속도만 있을 것이라고 예상할 수 있는데,
# 역시 이와 같은 결과를 위의 그래프에서 확인해 볼 수 있다.
```

```python
f(10) # R = 10kpc 인 경우
```

```python
f(12) # R = 12kpc 인 경우
```

```python
# 위의 두 그래프에서, 은경 0도를 기준으로 왼쪽은 은경이 양수, 오른쪽은 은경이 음수인 지점을 나타낸다. 즉 그래프의 왼쪽은
# 우리가 진행하는 방향에 있는 항성들, 그래프의 오른쪽은 우리의 진행 방향 반대쪽에 있는 항성들을 나타내고 있다.
# 우리가 진행하는 방향에 있는 항성(그래프 왼쪽)에 대해서 시선 속도가 음수로 나타나고 있는데, 이는 우리가 상대적으로 그 항성들에 대해서
# 가까워지고 있다는 사실에 부합한다. 그래프 오른쪽에 대해서도 같은 논의를 할 수 있다.
# 그리고 은경이 0도, 180도(혹은 -180도) 에 있는 천체들에 대해서 시선 속도가 0으로 나타나는데, 예상과 일치하는 결과이다.
```
