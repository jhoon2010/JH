```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import special
```

```python
k = np.linspace(0, 1, 100)
```

```python
# K(k), k: 0 -> 1
```

```python
plt.plot(k, scipy.special.ellipk(k))
```

```python
# E(k), k: 0 -> 1
```

```python
plt.plot(k, scipy.special.ellipe(k))
```

```python
# F(p, k), p: 0 -> pi/2, k: 0 -> 1 
```

```python
k = np.arange(0, 1, 0.1)
p = np.arange(0, np.pi/2, 0.1)
K, P = np.meshgrid(k, p)
Z = scipy.special.ellipkinc(P, K)

fig = plt.figure()
ax = fig.gca(projection = '3d')
surf = ax.plot_surface(K, P, Z)
```

```python
# E(p, k), p: 0 -> pi/2, k: 0 -> 1
```

```python
Z = scipy.special.ellipeinc(P,K)
fig = plt.figure()
ax = fig.gca(projection = '3d')
surf = ax.plot_surface(K, P, Z)
```
