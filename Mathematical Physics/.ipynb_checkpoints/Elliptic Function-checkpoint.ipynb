```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import special
```

```python
t = np.linspace(0, 1, 10)
```

```python
plt.plot(t, scipy.special.ellipk(t))
```

```python
plt.plot(t, scipy.special.ellipe(t))
```

```python
phi = np.linspace(0, np.pi / 2, 10)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
phi, t = np.meshgrid(phi, t)
result = scipy.special.ellipkinc(phi, t)
# plt.contour(result)
print(result)
```
