```python
def psi(x, t):
    F = np.exp(complex(0, - 4 * np.sqrt(x))) / (np.cos(2 * 4 * np.sqrt(1 + x)) - complex(0, 0.5 * (np.sqrt(x / (1 + x)) + np.sqrt(1 + x) / x) * np.sin(2 * 4 * np.sqrt(x + 1))))
    C = (np.sin(4 * np.sqrt(x + 1)) + complex(0, np.sqrt(x / (1 + x)) * np.cos(4 * np.sqrt(x + 1)))) * F
    D = (np.cos(4 * np.sqrt(x + 1)) - complex(0, np.sqrt(x / (1 + x)) * np.sin(4 * np.sqrt(x + 1)))) * F
    return C * np.cos(4 * np.sqrt(x + 1) * t) + D * np.sin(4 * np.sqrt(x + 1) * t)
```

```python
t = np.linspace(-1,1,1000)
plt.plot(t, psi(0.39, t).real)
```

```python
plt.plot(t, psi(1.47, t).real)
```

```python
plt.plot(t, psi(2.86, t).real)
```

```python
plt.plot(t, psi(4.55, t).real)
```

```python
def Tb(x):
    return (1 + 16 * x)**(-1)
```

```python
x = np.linspace(0, 1, 10000)
plt.plot(x, Ta(x))
plt.plot(1, Tb(1),'ro')
y = np.linspace(1, 5, 10000)
plt.plot(y, Tc(y))
```

```python
def psi_in(x,t):
    C = (np.exp(np.complex(0, - 4 * (np.sqrt(x) + np.sqrt(x - 1)))) * 4 * (np.sqrt(x) + np.sqrt(x - 1)) * 4 * np.sqrt(x)) / (np.complex(0, -(32 * x - 16)) * np.sin(8 * np.sqrt(x - 1)) + 2 * 16 * np.sqrt(x**2 - x) * np.cos(8 * np.sqrt(x - 1)))
    D = (np.exp(np.complex(0, - 4 * (np.sqrt(x) - np.sqrt(x - 1)))) * 4 * (np.sqrt(x) - np.sqrt(x - 1)) * 4 * np.sqrt(x)) / (np.complex(0, +(32 * x - 16)) * np.sin(8 * np.sqrt(x - 1)) - 2 * 16 * np.sqrt(x**2 - x) * np.cos(8 * np.sqrt(x - 1)))
    return C * np.cos(4 * t * np.sqrt(x - 1)) + C * complex(0, 1) * np.sin(4 * t * np.sqrt(x - 1)) + D * np.cos(4 * t * np.sqrt(x - 1)) - D * complex(0, 1) * np.sin(4 * t * np.sqrt(x - 1))
```
