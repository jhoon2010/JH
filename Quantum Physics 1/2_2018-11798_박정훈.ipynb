```python
# x : x/a 단위로
# t : tau 단위로

a = 1
```

```python
def psi(x,t,n):
    return (-8 * 15**(0.5) / (np.pi)**3) / (n**3) * np.sqrt(2 / a)* np.sin(n * np.pi / 2) * np.cos(n * np.pi * x) * np.exp(-complex(0,(n**2) * t))
```

```python
def psicon(x,t,n):
    return (-8 * 15**(0.5) / (np.pi)**3) / (n**3) * np.sqrt(2 / a)* np.sin(n * np.pi / 2) * np.cos(n * np.pi * x) * np.exp(complex(0,(n**2) * t))
```

```python
def psi_square(x,t):
    return ((psicon(x,t,1) + psicon(x,t,3) + psicon(x,t,5)) * (psi(x,t,1) + psi(x,t,3) + psi(x,t,5)))
```

```python
x = np.linspace(-0.5, 0.5, 10000)
plt.plot(x, psi_square(x,0),'r')
plt.xlim(-0.5,0.5)
plt.ylim(0,2.5)
plt.title('t = 0')
```

```python
plt.plot(x, psi_square(x,1),'r')
plt.xlim(-0.5,0.5)
plt.ylim(0,2.5)
plt.title('t = tau')
```

```python
plt.plot(x, psi_square(x,2),'r')
plt.xlim(-0.5,0.5)
plt.ylim(0,2.5)
plt.title('t = 2 * tau')
```

```python
x, beta, hbar = symbols("x, beta, hbar")
```

```python
# ground state
psi_0 = beta**(0.5) * pi**(-0.25) * exp(-(beta)**2 * (x**2) / 2)
```

```python
def Raise(psi, n):
    for i in range(n):
        psi = ((beta * x * psi) - (diff(psi, x)) / beta) / sqrt(2)
        psi = psi / sqrt(i + 1)
    return psi.simplify()
```

```python
Raise(psi_0,0)
```

```python
Raise(psi_0,1)
```

```python
Raise(psi_0,2)
```

```python
Raise(psi_0,3)
```

# Problem 7 -(f)

```python
def integral(func, init, fin):
    point = init
    dt = 0.0001
    SUM = 0
    while(point < fin):
        SUM += func(point) * dt
        point += dt
    return SUM
```

```python
def psi__0(t):
    return np.exp(-(t**2))
```

```python
a0 = integral(psi__0, -1, 1)
print(a0)
1 - a0 / np.sqrt(np.pi)
```

```python
def psi__1(t):
    return 2 * t**2 * np.exp(-(t**2))
```

```python
a1 = integral(psi__1, -np.sqrt(3), np.sqrt(3))
print(a1)
1 - a1 / np.sqrt(np.pi)
```

```python
def psi__2(t):
    return 1 / 2 * (-1 + 2 * t**2)**2 * np.exp(-(t**2))
```
